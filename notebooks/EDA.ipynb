```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../vehicles.csv')
df.head()
```

```python
df.info()
df.describe()
df.columns.tolist()
```

```python
px.histogram(df, x='odometer', nbins=50, title='Histograma - odometer')
```

```python
px.scatter(df, x='odometer', y='price', color='paint_color', title='Preço vs Odometer')
```
